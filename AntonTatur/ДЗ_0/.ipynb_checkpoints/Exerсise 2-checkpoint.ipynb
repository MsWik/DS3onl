```python
#Нарисовать линейный график зависимости цельсий от фаренгейта и второй график с ошибкой. Способ 1

import matplotlib.pyplot as plt

f=[]
c1=[]
c2=[]


def c_calc(f):
    return (5 / 9) * (f - 32)

def c_calc2(f):
    return (5 / 9) * (f)

for x in range(-1,100,1):
    f.append(x)

for x in range(len(f)):
    c1.append (c_calc(f[x]))
    c2.append(c_calc2(f[x]))

plt.plot(f,c1)
plt.plot(f,c2)

plt.show()
```

```python
#Нарисовать линейный график зависимости цельсий от фаренгейта и второй график с ошибкой. Способ 2


import matplotlib.pyplot as plt

c_calc=lambda f: (5 / 9) * (f - 32)
c_calc2 = lambda f: (5 / 9) * (f)
f=[c for c in range(100)]

c1_list=[c_calc(x) for x in f]
c2_list=[c_calc2(x) for x in f]

plt.plot(f,c1_list)
plt.plot(f,c2_list)

plt.show()
```
