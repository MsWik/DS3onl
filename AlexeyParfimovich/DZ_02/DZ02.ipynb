Задача №1.	
Напишите программу, которая рассчитывает расход топлива маршрута.
На вход подается длина маршрута и расход топлива на 100 километров. 
Используйте try catch (проверка что пользователь ввел число>0). 

```python
print('Расчет расхода топлива транспортным средством (ТС) на маршрут')
try:
    print('Введите протяженность маршрута ТС, км:')
    path_length = float(input())
    assert path_length > 0, "Протяженность маршрута должна быть больше 0"

    print('Введите средний расход ТС на 100 км пути, литров:')
    avg_consumption = float(input())
    assert avg_consumption > 0, "Расход топлива должен быть больше 0"

    path_cosumption = path_length * avg_consumption / 100

    print(f"Расход топлива на {path_length} км составит {path_cosumption:.3f} л, при среднем расходе {avg_consumption} л/100км.")
except ValueError as e:
    print('Ведено нечисловое значение:', e)
except AssertionError as e:
    print('Ведено некорректное значение:', e)
```

Задача №2.
Напишите программу, которая ищет корни квадратного уравнения вида a*x^2+b*x+c. 
На вход программы подаются a,b,c. Результатом работы программы – действительные корни или сообщение что корней нет. 

```python
print('Нахождение действительных корней квадратного уравнения вида a*x^2 + b*x + c = 0')
try:
    print('Ведите через пробел значения коэффициентов a, b и c :')
    a, b, c = map(float, input().split())

    if a == 0 or b == 0:
        print('Один из коэффициентов равен 0. Квадратное уравнение неполное.')
        if a == 0 and b == 0:
            print('Корней нет')
        elif a == 0 and b != 0:
            print(f'x = {-c / b}')
        else:
            x2 = -c / a
            if x2 < 0:
                print('Корней нет')
            else:
                print(f'x = {x2 ** 0.5}')
    else:
        discr = b ** 2 - 4 * a * c
        print(f'Дискриминант = {discr}')

        if discr < 0:
            print('Корней нет')
        elif discr == 0:
            print('Уравнение имеет один корень:')
            print(f'x = {-b / (2 * a)}')
        else:
            print('Уравнение имеет два корня:')
            print(f'x1 = {(-b + discr ** 0.5) / (2 * a)}')
            print(f'x2 = {(-b - discr ** 0.5) / (2 * a)}')

except ValueError as e:
    print('Ведено нечисловое значение:', e)
```

Задача №3
Напишите программу, находящую наибольшее и наименьшее значение функции y = -26*x^2+25*x-9 на отрезке [-5,5]. 
Для решения используйте while и произвольный шаг (например 0.1). 

```python
def get_function_value(arg):
    return -26 * arg**2 + 25 * arg - 9

print('Нахождение наибольшего и наименьшего значений функции вида "y = -26*x^2+25*x-9" на заданном интервале значений аргумента')
try:
    print('Ведите через пробел начальное, конечное значения и шаг дискретизации аргумента функции:')
    x_init, x_final, sampl_step = map(float, input().split())

    assert x_init < x_final, "начальное значение меньше конечного"

    assert x_final - x_init > sampl_step, "шаг дискретизации слишком большой"

    x = x_min = x_max = x_init
    y_min = y_max = get_function_value(x)

    while x <= x_final:
        y = get_function_value(x)
        if y < y_min: 
            y_min = y
            x_min = x
        elif y > y_max: 
            y_max = y
            x_max = x
        x += sampl_step
    
    print(f'Функция принимает минимальное значение y = {y_min:.3f} для аргумента x = {x_min:.3f}')
    print(f'Функция принимает максимальное значение y = {y_max:.3f} для аргумента x = {x_max:.3f}')

except ValueError as e:
    print('Ведено нечисловое значение:', e)
except AssertionError as e:
    print('Ведены некорректные значения интервала:', e)
```
